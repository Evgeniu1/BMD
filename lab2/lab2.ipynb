```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("diagram")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(x, y, 'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, 'green')
axes.set_xlabel('X_LABEL')
axes.set_ylabel('Y_LABEL')
axes.set_title('Graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Axes_1')


axes2.plot(y, x, 'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Axes_2')
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, '*')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('AXES')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")

ax.legend()
fig.savefig("GRAPH.jpg")
```
