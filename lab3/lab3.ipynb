```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = "/Users/evgeniu/Desktop/BMD/lab3/planets.csv"
```

```python
planet = pd.read_csv(data, encoding="utf-8")
```

```python
planet.head()
```

```python
sns.distplot(planet['mass'])
```

```python
sns.jointplot(x = "distance", y = "mass", data = planet)
```

```python
sns.pairplot(planet)
```

```python
sns.pairplot(planet, hue = 'number')
```

```python
sns.barplot(x = "mass", y = "year", data = planet)
```

```python
sns.barplot(x = 'orbital_period', y = 'year', data = planet, estimator = np.median)
```

```python
sns.boxplot(x = 'year', y = 'number', data = planet, palette = 'rainbow_r')
```

```python
sns.stripplot(x = 'number', y = 'distance', data = planet)
```

```python
sns.heatmap(planet.corr())
```

```python
g = sns.FacetGrid(planet, col = 'method', row = 'number')
g = g.map(plt.hist, 'mass')
```
